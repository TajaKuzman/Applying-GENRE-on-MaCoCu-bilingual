```python
import re
import json
import pandas as pd
```

```python
corpus=pd.read_csv("Macocu-sl-en-doc-format.csv", sep="\t", index_col = 0)
corpus
```

```python
# Calculate ratio of punctuations per words

def paragraph_punct_ratio(text):
    token_re=re.compile(r'\w+|\S',re.UNICODE)
    tokens=token_re.findall(text)
    punct=len([e for e in tokens if e in '.;,!?:'])
    ratio = punct/len(tokens)
    return ratio

corpus["punct_ratio"] = corpus.en_doc.apply(paragraph_punct_ratio)

corpus.head(3)
```

```python
corpus.punct_ratio.describe()
```

```python
# With the ratio < 0.015 we catch non-textual texts without any punctuation
for i in corpus.query("punct_ratio < 0.015").en_doc.to_list()[:5]:
	print(i)
```

```python
# With ratio above 0.2, we catch non-textual texts with a lot of punctuations
for i in corpus.query("punct_ratio > 0.20").en_doc.to_list()[:5]:
	print(i)
```

```python
# Filter the corpus by using only instances with ratio between 0.015 and 0.3
corpus = corpus.query("punct_ratio >= 0.015 & punct_ratio <= 0.2")

corpus.head()
```

```python
corpus.describe()
```

We remove 1474 texts and are left with 101,807 texts.

```python
# Save the new file
corpus.to_csv("Macocu-sl-en-doc-format-filtered.csv", sep= "\t")
```

```python
# Inspect whether this solved any issues with non-textual texts
non_textual_index = [150635, 2050301, 1738274, 2754062, 84675]

discarded_counter = 0

remaining_index = []

for i in non_textual_index:
	try:
		print(corpus.loc[i].en_doc)
		print(corpus.loc[i].punct_ratio)
		remaining_index.append(i)

	except KeyError:
		print(f"{i} not in final corpus.")
		discarded_counter += 1

print(discarded_counter)
print(remaining_index)
```

By filtering based on punctuation, we got rid of 2 out of 5 remaining non-textual texts.
